## North Dakota Linear Regression

```
#installing packages to read pingouin
!pip install pingouin


#Import standard libraries
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import pingouin
from scipy.stats import chi2_contingency
import matplotlib.colors as mcolors
import statsmodels.formula.api as sm
```

```
# NORTH DATOKA LINEAR REGRESSION


#read in north dakota data
north_dakota = pd.read_csv('data by state/north_dakota.csv')
```

```
#linear regression using homosexual from FBI Crime Data & Google Search Terms

nd_homosexual_relationship_reg = ' homosexual_relationship ~ domestic_violence + sexual_violence + handgun + abuse + rape'

#model output
nd_homosexual_model = sm.ols(formula = nd_homosexual_relationship_reg, data = north_dakota)
nd_homosexual_model_fit = nd_homosexual_model.fit()
print(nd_homosexual_model_fit.summary())
```

```
#linear regression using relationship status from FBI Crime & Google Search Terms
nd_relationship_reg = 'boyfriend_girlfriend ~ domestic_violence + sexual_violence + handgun + abuse+ rape'

#model output
nd_relationship_model = sm.ols(formula = nd_relationship_reg, data = north_dakota)
nd_relationship_model_fit = nd_relationship_model.fit()
print(nd_relationship_model_fit.summary())
```

```
#linear regression using ex-spouse from FBI Crime & Google Search Terms
nd_exspouse_reg = ' ex_spouse ~ domestic_violence + sexual_violence + handgun + abuse + rape'

#model output
nd_exspouse_model = sm.ols(formula = nd_exspouse_reg, data = north_dakota)
nd_exspouse_model_fit = nd_exspouse_model.fit()
print(nd_exspouse_model_fit.summary())
```

```
#linear regression using spouse from FBI Crime & Google Search Terms
nd_spouse_reg = 'spouse ~ domestic_violence + sexual_violence + handgun + abuse + rape'

#model output
nd_spouse_model = sm.ols(formula = nd_spouse_reg, data = north_dakota)
nd_spouse_model_fit = nd_spouse_model.fit()
print(nd_spouse_model_fit.summary())
```
