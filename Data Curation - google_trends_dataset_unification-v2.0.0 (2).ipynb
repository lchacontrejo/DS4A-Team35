```python
import pandas as pd
```

```python
#Steps
## define URLs
## import csv files as dataframes
## changes column name to standardize
## replacing <1 in'sexual_violence' term to 0
# transform ND_df['sexual_violence'] from object to int
## transforming month column object %Y%m as date type
## extract year from the date 
## group data by year and reset index
## create state and state_abbrv columns
## concat dataframes
## removing rows with 2022 year data
## export dataframe as csv
```

```python
#Google trends Interest over time data
#define local URL
TX_URL_IOT = 'google_trends_data/multiTimeline-TX-webs.csv'
TN_URL_IOT = 'google_trends_data/multiTimeline-TN-webs.csv'
VT_URL_IOT = 'google_trends_data/multiTimeline-Vt-webs.csv'
IL_URL_IOT = 'google_trends_data/multiTimeline-IL-webs.csv'
AK_URL_IOT = 'google_trends_data/multiTimeline-AK-webs.csv'
ND_URL_IOT = 'google_trends_data/multiTimeline-ND-webs.csv'
MN_URL_IOT = 'google_trends_data/multiTimeline-MN-webs.csv'
MI_URL_IOT = 'google_trends_data/multiTimeline-MI-webs.csv' 
```

```python
## import csv files as dataframes using pandas
TX_df = pd.read_csv(TX_URL_IOT, skiprows=2)
IL_df = pd.read_csv(IL_URL_IOT, skiprows=2)
AK_df = pd.read_csv(AK_URL_IOT, skiprows=2)
ND_df = pd.read_csv(ND_URL_IOT, skiprows=2)
MN_df = pd.read_csv(MN_URL_IOT, skiprows=2)
MI_df = pd.read_csv(MI_URL_IOT, skiprows=2)
TN_df = pd.read_csv(TN_URL_IOT, skiprows=2)
VT_df = pd.read_csv(VT_URL_IOT, skiprows=2)
```

```python
# dataframes variables to validate
# TX_df
# IL_df
# AK_df
# ND_df
# MN_df
# MI_df
# TN_df
# VT_df
TX_df.columns
```

```python
##changes column name to standardize
columns_dict_TX = {'"domestic violence": (Texas)':'domestic_violence',
                    '"handgun": (Texas)':'handgun',
                    '"rape": (Texas)':'rape',
                    '"abuse": (Texas)':'abuse',
                    '"sexual violence": (Texas)':'sexual_violence',
                    'Month': 'month'
                    }
    

TX_df = TX_df.rename(columns=columns_dict_TX)
TX_df 


columns_dict_IL = {'"domestic violence": (Illinois)':'domestic_violence',
                    '"handgun": (Illinois)':'handgun',
                    '"rape": (Illinois)':'rape',
                    '"abuse": (Illinois)':'abuse',
                    '"sexual violence": (Illinois)':'sexual_violence',
                    'Month': 'month'
                    }
    

IL_df = IL_df.rename(columns=columns_dict_IL)
IL_df
columns_dict_IL = {'"domestic violence": (Illinois)':'domestic_violence',
                    '"handgun": (Illinois)':'handgun',
                    '"rape": (Illinois)':'rape',
                    '"abuse": (Illinois)':'abuse',
                    '"sexual violence": (Illinois)':'sexual_violence',
                    'Month': 'month'
                    }
    

IL_df = IL_df.rename(columns=columns_dict_IL)
IL_df

columns_dict_AK = {'"domestic violence": (Alaska)':'domestic_violence',
                    '"handgun": (Alaska)':'handgun',
                    '"rape": (Alaska)':'rape',
                    '"abuse": (Alaska)':'abuse',
                    '"sexual violence": (Alaska)':'sexual_violence',
                    'Month': 'month'
                    }
    

AK_df = AK_df.rename(columns=columns_dict_AK)
AK_df

columns_dict_MI = {'"domestic violence": (Michigan)':'domestic_violence',
                    '"handgun": (Michigan)':'handgun',
                    '"rape": (Michigan)':'rape',
                    '"abuse": (Michigan)':'abuse',
                    '"sexual violence": (Michigan)':'sexual_violence',
                    'Month': 'month'
                    }
    

MI_df = MI_df.rename(columns=columns_dict_MI)
MI_df

columns_dict_MN = {'"domestic violence": (Minnesota)':'domestic_violence',
                    '"handgun": (Minnesota)':'handgun',
                    '"rape": (Minnesota)':'rape',
                    '"abuse": (Minnesota)':'abuse',
                    '"sexual violence": (Minnesota)':'sexual_violence',
                    'Month': 'month'
                    }
    

MN_df = MN_df.rename(columns=columns_dict_MN)
MN_df

columns_dict_ND = {'"domestic violence": (North Dakota)':'domestic_violence',
                    '"handgun": (North Dakota)':'handgun',
                    '"rape": (North Dakota)':'rape',
                    '"abuse": (North Dakota)':'abuse',
                    '"sexual violence": (North Dakota)':'sexual_violence',
                    'Month': 'month'
                    }
    

ND_df = ND_df.rename(columns=columns_dict_ND)
ND_df

columns_dict_TN = {'"domestic violence": (Tennessee)':'domestic_violence',
                    '"handgun": (Tennessee)':'handgun',
                    '"rape": (Tennessee)':'rape',
                    '"abuse": (Tennessee)':'abuse',
                    '"sexual violence": (Tennessee)':'sexual_violence',
                    'Month': 'month'
                    }
    

TN_df = TN_df.rename(columns=columns_dict_TN)
TN_df

columns_dict_VT = {'"domestic violence": (Vermont)':'domestic_violence',
                    '"handgun": (Vermont)':'handgun',
                    '"rape": (Vermont)':'rape',
                    '"abuse": (Vermont)':'abuse',
                    '"sexual violence": (Vermont)':'sexual_violence',
                    'Month': 'month'
                    }
    

VT_df = VT_df.rename(columns=columns_dict_VT)
VT_df
```

```python
#transforming month column object %Y%m as date type
VT_df['date'] = pd.to_datetime(VT_df['month']) #, format="%Y%m"
TN_df['date'] = pd.to_datetime(TN_df['month'])
AK_df['date'] = pd.to_datetime(AK_df['month'])
IL_df['date'] = pd.to_datetime(MI_df['month'])
MI_df['date'] = pd.to_datetime(VT_df['month'])
MN_df['date'] = pd.to_datetime(MN_df['month'])
TX_df['date'] = pd.to_datetime(TX_df['month'])
ND_df['date'] = pd.to_datetime(ND_df['month'])
```

```python
# replacing <1 in'sexual_violence' term to 0
TX_df['sexual_violence'] = TX_df['sexual_violence'].replace('<1', '0')
VT_df['sexual_violence'] = VT_df['sexual_violence'].replace('<1', '0')
TN_df['sexual_violence'] = TN_df['sexual_violence'].replace('<1', '0')
AK_df['sexual_violence'] = AK_df['sexual_violence'].replace('<1', '0')
IL_df['sexual_violence'] = IL_df['sexual_violence'].replace('<1', '0')
MI_df['sexual_violence'] = MI_df['sexual_violence'].replace('<1', '0')
MN_df['sexual_violence'] = MN_df['sexual_violence'].replace('<1', '0')
ND_df['sexual_violence'] = ND_df['sexual_violence'].replace('<1', '0')
```

```python
# transform ND_df['sexual_violence'] from object to int
TX_df['sexual_violence'] = TX_df['sexual_violence'].astype('int64')
IL_df['sexual_violence'] = IL_df['sexual_violence'].astype('int64')
AK_df['sexual_violence'] = AK_df['sexual_violence'].astype('int64')
ND_df['sexual_violence'] = ND_df['sexual_violence'].astype('int64')
MN_df['sexual_violence'] = MN_df['sexual_violence'].astype('int64')
MI_df['sexual_violence'] = MI_df['sexual_violence'].astype('int64')
TN_df['sexual_violence'] = TN_df['sexual_violence'].astype('int64')
VT_df['sexual_violence'] = VT_df['sexual_violence'].astype('int64')
```

```python
TX_df.info()
```

```python
#extracting year from the date
VT_df['year'] = VT_df['date'].dt.year 
TN_df['year'] = TN_df['date'].dt.year
AK_df['year'] = AK_df['date'].dt.year
IL_df['year'] = IL_df['date'].dt.year
MI_df['year'] = MI_df['date'].dt.year
MN_df['year'] = MN_df['date'].dt.year
TX_df['year'] = TX_df['date'].dt.year
ND_df['year'] = ND_df['date'].dt.year
MN_df
```

```python
##group data by year and reset index
VT_grouped = VT_df.groupby(['year']).sum().reset_index()
TN_grouped = TN_df.groupby(['year']).sum().reset_index()
AK_grouped = AK_df.groupby(['year']).sum().reset_index()
IL_grouped = IL_df.groupby(['year']).sum().reset_index()
MI_grouped = MI_df.groupby(['year']).sum().reset_index()
MN_grouped = MN_df.groupby(['year']).sum().reset_index()
TX_grouped = TX_df.groupby(['year']).sum().reset_index()
ND_grouped = ND_df.groupby(['year']).sum().reset_index()
AK_grouped
```

```python
## create state and state_abbrv columns

##Create State name column
TX_grouped['state'] = 'Texas'
IL_grouped['state'] = 'Illinois'
AK_grouped['state'] = 'Alaska'
ND_grouped['state'] = 'North Dakota'
MN_grouped['state'] = 'Minnesota'
MI_grouped['state'] = 'Michigan'
TN_grouped['state'] = 'Tennessee'
VT_grouped['state'] = 'Vermont'

##Create State abbreviation column
TX_grouped['state_abbrv'] = 'TX'
IL_grouped['state_abbrv'] = 'IL'
AK_grouped['state_abbrv'] = 'AK'
ND_grouped['state_abbrv'] = 'ND'
MN_grouped['state_abbrv'] = 'MN'
MI_grouped['state_abbrv'] = 'MI'
TN_grouped['state_abbrv'] = 'TN'
VT_grouped['state_abbrv'] = 'VT'

TN_grouped
```

```python
# concatenating dfataframes along rows
google_trends_df = pd.concat([TX_grouped, 
                              IL_grouped, 
                              AK_grouped, 
                              ND_grouped, 
                              MN_grouped, 
                              MI_grouped, 
                              TN_grouped, 
                              VT_grouped],  axis=0).reset_index() #join='inner',
google_trends_df = google_trends_df.drop(columns='index')
google_trends_df
```

```python
index_2022 = google_trends_df[google_trends_df['year'] == 2022].index
google_trends_df = google_trends_df.drop(index_2022)
#google_trends_df[google_trends_df['year'] == 2022]
```

```python
## export dataframe as csv
google_trends_df.to_csv('google_trends_data/clean_google_trends_data_v2.csv', index=False, encoding="utf-8")
```
